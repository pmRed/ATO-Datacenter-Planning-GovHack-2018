```python
import pandas as pd

datadir = "../data/"
!tree -L 3 "../data/"
```

# Combine the ATO Data

```python
ato_abs = pd.read_csv(datadir+"/ato/atoabsgovhack2018_abs.csv")
ato_abs = ato_abs.loc[ato_abs["Income year"] == 2016]
ato_abs.set_index("Postcode", inplace=True)

ato_absseifa = pd.read_csv(datadir+"/ato/atoabsgovhack2018_absseifa.csv")
ato_absseifa = ato_absseifa.loc[ato_absseifa["Year"] == 2016]
ato_absseifa.set_index("Postal Area (POA) Code", inplace=True)

ato_ato = pd.read_csv(datadir+"ato/atoabsgovhack2018_ato.csv")
ato_ato = ato_ato.loc[ato_ato["Income year"] == 2016]
ato_ato.set_index("Postcode", inplace=True)

ato_helpcentre = pd.read_csv(datadir+"ato/atoabsgovhack2018_helpcentre.csv")
ato_helpcentre.set_index("Post Code", inplace=True)

ato = pd.concat([ato_abs, ato_absseifa, ato_ato, ato_helpcentre], axis=1)
```

# Combine the Census Data

```python
census_G43A = pd.read_csv(datadir+"census/gcppoa/2016Census_G43A_AUS_POA.csv")
census_G43A["POA_CODE_2016"] = census_G43A["POA_CODE_2016"].apply(lambda x: int(x[3:]))
census_G43A.set_index("POA_CODE_2016", inplace=True)

census_G43B = pd.read_csv(datadir+"census/gcppoa/2016Census_G43B_AUS_POA.csv")
census_G43B["POA_CODE_2016"] = census_G43B["POA_CODE_2016"].apply(lambda x: int(x[3:]))
census_G43B.set_index("POA_CODE_2016", inplace=True)
census = pd.concat([census_G43A, census_G43B], axis=1)
census
```

# Combine the Census and ATO Data

```python
ez2data = pd.concat([census, ato], axis=1)
ez2data.to_csv("ez2munge.csv")
```
